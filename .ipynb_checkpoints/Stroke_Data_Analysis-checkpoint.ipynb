```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading Data
df = pd.read_csv('stroke_data.csv')
df.head()
```

```python
df.columns
```

```python
# Checking last three rows
df.tail()
```

```python
# Checking data types
df.info()
df.dtypes
```

```python
# Cleaning data type 
df['age'] = df['age'].astype(int)
df['age'].dtypes
df['hypertension'] = df['hypertension'].astype(bool)
df['heart_disease'] = df['heart_disease'].astype(bool)
df['stroke'] = df['stroke'].astype(bool)
df.dtypes
```

```python
#Changing all characters to lowercase
df.columns= df.columns.str.lower()
df.columns
```

```python
# Checking the number of rows and columns of the dataset
df.shape
```

```python
# Checking for missingness
df.isna()
```

```python
# Checking sum of missingness in the columns
df.isna().sum()
```

```python
# Checking if any of the columns has  missing values and returns True if present and False if not
df.isna().any()
```

```python
# To check the summary statistics of the data
df.describe()
```

```python
# Data Cleaning
df.nunique()
```

```python
#Removing trailing whitespace
df.columns = df.columns.str.strip()
df.columns
```

```python
# We can have a look at the proportion of missing data
df['bmi'].isnull().mean()
```

```python
df.head()
```

```python
df.describe()
```

```python
# BMI  column imputation with mean value
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
df
# We used mean value because:
    #1. The variable is numerical
    #2. The variable is randomly distributed across the dataset
    #3. Filling missing values with the mean is a simple and computationally efficient method. It is easy to implement, and it doesn't introduce complexities in the data
```

```python
df.head()
df.isna().any()
```

```python
# Checking and handling duplicates
df.duplicated()
# Check for duplicates sum per  in the dataframe
df.duplicated().sum()
```

```python
# Detecting and handling Outliers
# Using boxplot to visualize Outliers

# We decided not to drop the outliers

# Dropped the id column to better visualize boxplot
df= df.drop(df.columns[0],axis=1)
sns.boxplot(data=df)
```

```python
# Feature Scaling
from sklearn import preprocessing
import numpy as np

a=np.random.random((1,4))
a=a*20
print("df=",a)

# normalize the df attributes
normalized = preprocessing.normalize(a)
print("Normalized_df=", normalized)
```

```python
x_array = np.array([0,2,3,5,6,7,4,8,7,6,10])
normalized_arr = preprocessing.normalize([x_array])
print(x_array)
print(normalized_arr)
```

```python
# Standardization
# Import standard scaler from sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Slice the data using the columns age, RM
sliced_data = df[['age','avg_glucose_level','bmi']]
print(sliced_data)

# Once we have the data ready, we can use the StandardScaler() class and
# its methods (from sklearn library) to standardize the data:
sliced_data2= scaler.fit_transform(sliced_data)
sliced_data2
```

```python
# As you can see, the above code returned an array,
# so the last step would be to convert it to dataframe
standardized_df = pd.DataFrame(sliced_data2, columns=sliced_data.columns)
print(standardized_df)
```

```python
gender_stroke_counts = df.groupby(['gender', 'stroke']).size().unstack(fill_value=0)
sns.countplot(x='gender', hue='stroke', data=df)
plt.title('Distribution of Strokes based on Gender')
plt.show()
```

```python
work_type_stroke_counts = df.groupby(['work_type', 'stroke']).size().unstack(fill_value=0)
sns.countplot(x='work_type', hue='stroke', data=df)
plt.title('Distribution of Strokes based on Gender')
plt.show()
```

```python
df['smoking_status'].value_counts()
```

```python
# corecting the pe
old_spelling = 'smokes'
new_spelling = 'smokers'

df['smoking_status'] = df['smoking_status'].replace(old_spelling, new_spelling)
```

```python
smoking_status_stroke_counts = df.groupby(['smoking_status', 'stroke']).size().unstack(fill_value=0)
sns.countplot(x='smoking_status', hue='stroke', data=df)
plt.title('Distribution of Strokes based on Smoking Status')
plt.show()
```

```python
correlation_matrix = df[['gender', 'work_type', 'smoking_status', 'stroke']].corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
'''
Based on the heatmap provided, there appears to be a moderate negative correlation between work_type and stroke.
The correlation coefficient between these two variables is -0.4, 
which indicates that as the value of work_type increases, the value of stroke tends to decrease.

'''
```
